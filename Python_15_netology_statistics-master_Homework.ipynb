```python
# Считаем стандартное отклонение
def std_deviation(item):
    diffs = 0
    exp_value = sum(item) / len(item)
    for x in item:
        diffs += (x - exp_value)**(2)
    return (diffs/len(item))**(0.5)
   
print(std_deviation(freedom))
```

```python
# Считаем дисперсию
print(std_deviation(freedom)**2)
```

Почему нельзя найти моду?

```python
# Потому, что у каждая оценка проставлена по одному разу
# dict_

# Но почему-то такой вариант даёт моду = 5
dict_ = {}
free = df['Freedom']
for x in free:
    if x not in dict_:
        dict_[x] = 1
    else:
        dict_[x] += 1
    
mp_freq = 0
mp = 0

for k, v in dict_.items():
    if mp_freq < v:
        mp_freq = v 
        mp = k
print('Значение моды: ', mp, 'Количество встречаемости: ', mp_freq)
```

```python
avocado['date'] = pd.to_datetime(avocado['date'])
pred_type = 'organic'
avoc_pred = avocado[avocado.type == pred_type]
```

```python
territory = avoc_pred.groupby(avoc_pred.region)
```

```python
pred_region = 'LasVegas'
date_price = territory.get_group(pred_region)[['date', 'average_price']].reset_index(drop=True)
date_price.head()
```

```python
date_price.plot(x='date', y='average_price')
plt.title('Dynamic of price')
plt.ylabel('average_price')
plt.xlabel('date')
plt.grid(True)
```

```python
date_price = date_price.rename(columns={'date':'ds', 'average_price':'y'})
date_price = date_price.sort_values(by='ds')
amount_80 = int(len(date_price)*0.8)
train_price = date_price[:amount_80]
test_price = date_price[amount_80:]
```
