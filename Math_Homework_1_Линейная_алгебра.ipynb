
Повернуть на 130 градусов.

```
theta = 130 * 180/pi
TransfromArray = np.array([
  [cos(theta), sin(theta), 0],
  [-sin(theta), cos(theta), 0],
  [0, 0, 1]
])
```

```
A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]

A = np.array(A)
```

```
B = []
for point in A:
  new_point = np.dot(TransfromArray, np.transpose(point))
  B.append(new_point)
B = np.array(B).astype(int)
print(B)
```

```
plt.figure(figsize=(9,9))
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
x = B[:,0]
y = B[:,1]
plt.plot(x, y, color='g')
plt.ylim([-300, 300])
plt.xlim([-300, 300])
plt.legend(['new', 'old'])
plt.show()
```
