```python
# Задание 1
# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
# - оценка 2 и меньше - низкий рейтинг
# - оценка 4 и меньше - средний рейтинг
# - оценка 4.5 и 5 - высокий рейтинг
# Результат классификации запишите в столбец class
import pandas as pd
ratings = pd.read_csv('ratings.csv')
# ratings.head()
def classification(item):
    if item['rating'] <= 2:
        return 'low rating'
    if item['rating'] > 2 and item['rating'] <= 4:
        return 'middle rating'
    if item['rating'] >= 4.5 :
        return 'high rating'

ratings['class'] = ratings.apply(classification, axis=1)
ratings.head(20)
```

```python
# Задание 2
# Используем файл keywords.csv.
# Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность 
# определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ 
# пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
# Правила распределения по регионам Центр, Северо-Запад и Дальний Восток: 
# Результат классификации запишите в отдельный столбец region.
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

geoclass = pd.read_csv('keywords.csv')

def regions(item):
    for geo_region, cities in geo_data.items():
        for city in cities:
            if city in item['keyword']:
                return geo_region

    return 'Undefined'

geoclass['region'] = geoclass.apply(regions, axis=1)
geoclass[geoclass['region'].str.contains('Северо-Запад')].head(100)
```

```python
# В переменную years запишите список из всех годов с 1950 по 2010.
# Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. 
# Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
# для каждой строки пройдите по всем годам списка years
# если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
# если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
# Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
# Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга
import pandas as pd
import re
years = pd.read_csv('movies.csv')
rating = pd.read_csv('ratings.csv')
combine = pd.merge(years, rating, on = ['movieId'])

def production_year(dataframe):
    pattern = '[0-9]{4}'
    a = re.findall(pattern, dataframe)
    for valid_year in a:
        year_is_valid = 1950 <= int(valid_year) <= 2010
        if year_is_valid:
            return valid_year
    return 1900

combine['years'] = combine['title'].apply(production_year)
middle_rating = combine.groupby('years').mean()
middle_rating.sort_values(by='rating', ascending=False).head(20)
```
