```python
# Задание 1
# Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
# Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для
# данного значения userId.
import pandas as pd
ratings = pd.read_csv('ratings.csv')
df1 = ratings.groupby('userId').count()
new_df1 = df1[['movieId']]
new_df1.columns = ['rating_count']
new_df2 = ratings.merge(new_df1, how='left', on='userId')
new_df2.head()
```

```python
def groupby_function(data):
    for item in data['rating_count']:
        if item > 100:
            return data.timestamp.max() - data.timestamp.min()
new_df2.groupby('userId').apply(groupby_function).mean()
```

```python
geo_coordinates = pd.DataFrame(
    {
        'user': {  11 : {
        'coordinates_1': [12345.12345], 
        'coordinates_2': [23456.23456],
        'coordinates_3': [34567.34567],
    },
                   22 : {
        'coordinates_1': [23423.45689], 
        'coordinates_2': [57313.55567],
        'coordinates_3': [45787.09754],
    },
                   55 : {
        'coordinates_1': [55563.33467], 
        'coordinates_2': [11123.88754],
        'coordinates_3': [44667.55533],
    }
    
    }
   
    }
)

new_geo_coord = geo_coordinates.reset_index().rename({'index':'user_id', 'user':'coordinates'}, axis='columns')
new_geo_coord


def new_column(data):
    for item in data['coordinates']:
        for x, values in item.items():
            if data['user_id']:
                return values
            
new_geo_coord['x'] = new_geo_coord.apply(new_column, axis=1)
```
